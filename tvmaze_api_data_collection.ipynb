# Streaming Platform Content Strategy Using TVMaze API

## Project Overview
This project focuses on collecting TV show data from TVMaze API and converting the raw JSON responses into structured CSV files. The collected data will later be used to analyze content types, genres, languages, ratings, runtime, networks, and release trends.

## Business Context
As the founder of a streaming platform, I need data-driven insights to decide which types of content to promote, license, or invest in. Understanding genres, ratings, languages, and content availability can support better content strategy and user engagement.

## Problem Statement
The streaming platform needs to make better content acquisition and promotion decisions. Without data, it may invest in content types or genres that do not match audience preferences or quality indicators. By analyzing TV show genres, ratings, languages, networks, and release trends, the platform can identify which content areas are more valuable to prioritize.

## Research Questions
As the founder of a streaming platform, these questions will help guide content strategy and investment decisions:

1. Which genres have the highest average ratings and should be prioritized for content acquisition?
2. Which languages are most common in the available TV show catalog?
3. Does runtime have any relationship with average show rating?
4. Which networks produce the highest-rated shows?
5. How has TV show production changed over time based on premiere dates?
6. Which content types or genres should the platform prioritize based on both rating and availability?

## Data Source & Collection Method
The data was collected using API requests from the TVMaze public API. Since the Shows API is paginated, multiple pages were requested using the page parameter. The final dataset contains 750 TV show records. The API returned the data in JSON format, then it was converted into structured CSV files using Python and Pandas.

API endpoint used:
- Shows API: https://api.tvmaze.com/shows
